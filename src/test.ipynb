```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ==============================
# LOAD DATA
# ==============================
df = pd.read_csv("data/raw_reviews_20.csv")

# ==============================
# HITUNG JUMLAH REVIEW PER VERSI
# ==============================
counts = (
    df['reviewCreatedVersion']
    .value_counts()
    .sort_values(ascending=False)
)

print("\nJumlah review per versi:")
print(counts)

# ==============================
# BAR CHART
# ==============================
plt.figure(figsize=(14, 6))

bars = plt.bar(counts.index, counts.values)

# Label jumlah review
for bar in bars:
    yval = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width() / 2,
        yval,
        f"{int(yval)}",
        ha='center',
        va='bottom',
        fontsize=8
    )

plt.title("Jumlah Review per Versi")
plt.xlabel("Versi")
plt.ylabel("Jumlah Review")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# ==============================
# BAR CHART LOG SCALE
# ==============================
plt.figure(figsize=(14, 6))

bars = plt.bar(counts.index, counts.values)

for bar in bars:
    yval = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width() / 2,
        yval,
        f"{int(yval)}",
        ha='center',
        va='bottom',
        fontsize=8
    )

plt.yscale('log')

plt.title("Jumlah Review per Versi (Log Scale)")
plt.xlabel("Versi")
plt.ylabel("Jumlah Review (log scale)")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# ==============================
# BOXPLOT NORMAL
# ==============================
plt.figure(figsize=(7, 5))

plt.boxplot(counts.values)

plt.title("Boxplot Jumlah Review per Versi")
plt.ylabel("Jumlah Review")

plt.tight_layout()
plt.show()

# ==============================
# BOXPLOT LOG SCALE
# ==============================
plt.figure(figsize=(7, 5))

plt.boxplot(counts.values)

plt.yscale('log')

plt.title("Boxplot Jumlah Review per Versi (Log Scale)")
plt.ylabel("Jumlah Review (log scale)")

plt.tight_layout()
plt.show()

# ==============================
# HISTOGRAM NORMAL
# ==============================
plt.figure(figsize=(8, 5))

plt.hist(counts.values, bins=10)

plt.title("Distribusi Jumlah Review per Versi")
plt.xlabel("Jumlah Review")
plt.ylabel("Frekuensi")

plt.tight_layout()
plt.show()

# ==============================
# HISTOGRAM LOG SCALE
# ==============================
plt.figure(figsize=(8, 5))

plt.hist(np.log10(counts.values), bins=8)

plt.title("Distribusi Log Jumlah Review per Versi")
plt.xlabel("log10(Jumlah Review)")
plt.ylabel("Frekuensi")

plt.tight_layout()
plt.show()

# ==============================
# HITUNG IQR LOG
# ==============================
log_counts = np.log10(counts.values)

Q1 = np.percentile(log_counts, 25)
Q3 = np.percentile(log_counts, 75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR

# kembali ke skala asli
threshold = 10 ** lower_bound

print("\n===== HASIL IQR LOG =====")
print(f"Q1            : {Q1:.3f}")
print(f"Q3            : {Q3:.3f}")
print(f"IQR           : {IQR:.3f}")
print(f"Lower Bound   : {lower_bound:.3f}")
print(f"Threshold asli: {threshold:.2f} review")

# ==============================
# DETEKSI OUTLIER RENDAH
# ==============================
outlier_versions = counts[counts < threshold]

print("\nVersi yang dianggap outlier rendah:")
print(outlier_versions)

# ==============================
# VERSI VALID
# ==============================
valid_versions = counts[counts >= threshold]

print("\nVersi yang digunakan:")
print(valid_versions)
```
